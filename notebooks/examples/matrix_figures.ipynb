```python
#plt.imshow(square4, cmap=plt.get_cmap('mycmap'))

def plot_matrix(mat, cmap=cmap):
    ax = plt.subplot(111)
    ax.imshow(mat, cmap=cmap)
    plt.axis('off')
    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            ax.text(j, i, \
                     mat[i, j], \
                     fontsize=22, fontweight='bold', \
                     color='white', ha='center', va='center')
    return ax 

plot_matrix(square4, cmap=gcmap)
#plt.savefig(HOME + "/Desktop/square_matrix.png", dpi=360, transparent=True, bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
p1 = plot_matrix(square4, cmap=gcmap)
# Create a Rectangle patch
rect = patches.Rectangle((-0.5, -0.5), 4, 1, linewidth=1, edgecolor='orange', facecolor='orange')
# Add the patch to the Axes
p1.add_patch(rect)

plt.savefig(HOME + "/Desktop/square_matrix_1.png", dpi=360, transparent=True, bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
def plot_matrix(mat, **kwargs):
    ax = plt.subplot(111)
    ax.imshow(mat, **kwargs)
    #plt.axis('off')
    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            ax.text(j, i, \
                     np.round(mat[i, j], 2), \
                     fontsize=12, fontweight='bold', \
                     color='white', ha='center', va='center')
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    return ax 

plt.figure(figsize=(10, 8))
plot_matrix(mat1, cmap=plt.get_cmap("viridis"), clim=(0, 1))
plt.savefig(HOME + "/Desktop/mbroad_1.png", dpi=360, transparent=True, bbox_inches='tight', pad_inches=0)
plt.show()
```
